# Data visualization

```python
plt.figure(figsize=(10, 6))
sns.histplot(marketing['cons.price.idx'], bins=30, kde=True)
plt.title('Distribution of Consumer Price Index vs Consumer Confidence Index')
plt.xlabel('Consumer Price Index')
```

```python
plt.figure(figsize=(10, 6))
fig, ax = plt.subplots()
sns.boxplot(x=marketing['cons.price.idx'], ax=ax)
plt.title('Boxplot of Consumer Price Index')
plt.xlabel('Consumer Price Index')
plt.show()
```
